```python
import pandas as pd, numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/yosantosh/MLOps/refs/heads/master/Class_3_ML_Pipeline/bank-full.csv",sep=";",index_col=False)
df.head()
```

```python
df.info()
```

```python
df.isna().sum(axis=0)
```

```python
df.dropna(axis=1)               #
df.drop_duplicates(inplace=True)
```

```python
X,y = df.iloc[:,:-1],df.iloc[:,-1]
```

```python
df_num = X.select_dtypes(np.number)
df_cat = X.select_dtypes(object)                          #
df_num.head(), df_cat.head()
```

```python
D = {i:df_cat[i].nunique() for i in df_cat.columns}
List = [i for i in D   if (D[i]/len(df)*100)<1.0]
List,D                                                             #
```

```python
oneHOt= OneHotEncoder(sparse_output=False)
df_cat= pd.DataFrame(oneHOt.fit_transform(df_cat[List]) , columns = oneHOt.get_feature_names_out())
df_cat.head()

```

```python
final_df = pd.concat([df_num,df_cat],axis=1)
```

```python
pca_pipeline = make_pipeline(StandardScaler(), PCA(n_components=0.99))
final_array= pca_pipeline.fit_transform(final_df)
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier



rf = RandomForestClassifier(n_estimators=130, min_samples_leaf=5)
rf.fit(final_array,y)
```

```python
pd.DataFrame(classification_report(y,rf.predict(final_array)))
```

```python
def preprocessing(df,test_size):

    df.dropna(axis=1)               #
    df.drop_duplicates()
    
    X,y = df.iloc[:,:-1],df.iloc[:,-1]

    df_num = X.select_dtypes(np.number)
    df_cat = X.select_dtypes(object)                          #

    D = {i:df_cat[i].nunique() for i in df_cat.columns}
    List = [i for i in D   if (D[i]/len(df)*100)<1.0]
                                            
    oneHOt= OneHotEncoder(sparse_output=False)
    df_cat= pd.DataFrame(oneHOt.fit_transform(df_cat[List]) , columns = oneHOt.get_feature_names_out())

    final_df = pd.concat([df_num,df_cat],axis=1)

    pca_pipeline = make_pipeline(StandardScaler(), PCA(n_components=0.99))
    final_array = pca_pipeline.fit_transform(final_df)

    X_train, X_test, y_train, y_test= train_test_split(final_array,y,test_size=test_size,random_state=3,shuffle=True)

    return [X_train, X_test, y_train, y_test]
    
```
